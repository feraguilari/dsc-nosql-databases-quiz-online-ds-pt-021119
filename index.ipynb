Write your answer below this line: 
_______________________________________________________________________________________________________________________________<br>
Fault tolerance is the property that enables a system to continue operating properly in the event of the failure of (or one or more faults within) some of its components. Resilient Distributed Databases are stored in different servers with backups in other servers, so if a server is to fail, the backup kicks in.
